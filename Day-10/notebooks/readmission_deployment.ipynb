### Create an “input_example.csv” for scoring (so you can demo one-command)
Since we don’t have real “new hospital data,” we’ll simulate it by extracting rows from DuckDB and dropping label.

Run this in a Day-10 notebook or in a one-off Jupyter cell anywhere:

```python
from pathlib import Path
import duckdb
import pandas as pd

project_root = Path.cwd().resolve()
while not (project_root / "Day-1").exists():
    project_root = project_root.parent

db_path = project_root / "Day-1" / "data" / "warehouse" / "day1.duckdb"
con = duckdb.connect(str(db_path))

df = con.execute("SELECT * FROM gold_diabetes_features_v1 LIMIT 2000").df()
con.close()

# pretend it's “new” data: keep encounter_id/person_id + all features, drop label
df_new = df.drop(columns=["label"])
out_path = project_root / "Day-10" / "data" / "input_example.csv"
out_path.parent.mkdir(parents=True, exist_ok=True)
df_new.to_csv(out_path, index=False)
print("Wrote:", out_path, "shape:", df_new.shape)
```
